```python
import re
instagram_tags =[]
def get_content(driver):
    
    # 1. í˜„ì¬ í˜ì´ì§€ì˜ HTML ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    # 2. ë³¸ë¬¸ë‚´ìš©
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content = ''
        
    # 3. ë³¸ë¬¸ ë‚´ìš©ì—ì„œ í•´ì‹œíƒœê·¸ ê°€ì ¸ì˜¤ê¸°(ì •ê·œí‘œí˜„ì‹ í™œìš©)
    tags = re.findall('#[A-Za-z0-9ê°€-í£]+', content)
    tag = ''.join(tags).replace("#"," ") # "#" ì œê±°
    tag_data = tag.split()
    
    for tag_one in tag_data:
        instagram_tags.append(tag_one)
        
    # content ë³€ìˆ˜ì˜ ë³¸ë¬¸ ë‚´ìš© ì¤‘ #ìœ¼ë¡œ ì‹œì‘í•˜ë©°, #ë’¤ì— ì—°ì†ëœ ë¬¸ì(ê³µë°±ì´ë‚˜ #, \ ê¸°í˜¸ê°€ ì•„ë‹Œ ê²½ìš°)ë¥¼ ëª¨ë‘ ì°¾ì•„ tags ë³€ìˆ˜ì— ì €ì¥
    
    # 4. ì‘ì„± ì¼ì ê°€ì ¸ì˜¤ê¸°
    try:
        #ì•ì—ì„œë¶€í„° 10ìë¦¬ ê¸€ì
        date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10]       
    except:
        date = ''
        
    # 5. ìœ„ì¹˜ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    try:
        place = soup.select('div.JF9hh')[0].text
    except:
        place = ''
    
    # 6. ìˆ˜ì§‘í•œ ì •ë³´ ì €ì¥í•˜ê¸°
    data = [content, date, place, tag_data]
    return data

get_content(driver)
```
